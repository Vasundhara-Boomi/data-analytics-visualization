```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Import data from the CSV file
df = pd.read_csv("population_literacy_region.csv")

# Convert the "Literacy Rate" column to numeric, coercing errors to NaN
df["Literacy Rate"] = pd.to_numeric(df["Literacy Rate"], errors='coerce')

# Remove rows with missing or NaN values in the "Literacy Rate" column
df = df.dropna(subset=["Literacy Rate"])

# Scatter plots for each region
regions = df["Region"].unique()
markers = ["D", "o", "s", "^"]  # Marker styles for different regions
colors = ["red", "blue", "green", "purple"]  # Marker colors for different regions

plt.figure(figsize=(10, 6))

for i, region in enumerate(regions):
    region_data = df[df["Region"] == region]
    plt.scatter(
        region_data["Population"],
        region_data["Literacy Rate"],
        label=region,
        marker=markers[i],
        s=np.sqrt(region_data["Literacy Rate"]),
        color=colors[i],
        alpha=0.7,
    )

plt.xlabel("Population")
plt.ylabel("Literacy Rate")
plt.title("Scatter Plot of Population vs. Literacy Rate by Region")
plt.legend()
plt.grid(True)

# Calculation of average literacy rate for each region
average_literacy_by_region = df.groupby("Region")["Literacy Rate"].mean()

# bar chart for average literacy rates
plt.figure(figsize=(8, 5))
average_literacy_by_region.plot(kind="bar", color="skyblue")
plt.xlabel("Region")
plt.ylabel("Average Literacy Rate")
plt.title("Average Literacy Rate by Region")
plt.xticks(rotation=0)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    "Name": ["Student 1", "Student 2", "Student 3", "Student 4", "Student 5"],
    "Number of Siblings": [2, 1, 3, 0, 2],
    "Father's Occupation": ["Engineer", "Doctor", "Teacher", "Business", "Lawyer"],
    "Mother's Occupation": ["Nurse", "Engineer", "Homemaker", "Business", "Teacher"],
    "Current GPA": [3.8, 4.0, 3.5, 3.2, 3.9],
    "Favorite Subject": ["Math", "Science", "History", "English", "Art"],
}

df = pd.DataFrame(data)

# Bar Chart - Family Background
plt.figure(figsize=(10, 6))
df["Number of Siblings"].plot(kind="bar", color="skyblue", alpha=0.7)
plt.xlabel("Student")
plt.ylabel("Number of Siblings")
plt.title("Number of Siblings among Students")
plt.xticks(np.arange(len(df)), df["Name"], rotation=0)
plt.tight_layout()
plt.show()

# Scatter Diagram - Class Performance
plt.figure(figsize=(10, 6))
plt.scatter(df["Favorite Subject"], df["Current GPA"], c="purple", marker="o", s=100)
plt.xlabel("Favorite Subject")
plt.ylabel("Current GPA")
plt.title("Class Performance vs. Favorite Subject")
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("company_sales_data.csv")
monthList  = df ['month_number'].tolist()

faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

plt.plot([],[],color='silver', label='face Cream', linewidth=5)
plt.plot([],[],color='pink', label='Face wash', linewidth=5)
plt.plot([],[],color='beige', label='Tooth paste', linewidth=5)
plt.plot([],[],color='mediumaquamarine', label='Bathing soap', linewidth=5)
plt.plot([],[],color='mediumpurple', label='Shampoo', linewidth=5)
plt.plot([],[],color='lavender', label='Moisturizer', linewidth=5)

plt.stackplot(monthList, faceCremSalesData, faceWashSalesData, toothPasteSalesData, 
              bathingsoapSalesData, shampooSalesData, moisturizerSalesData, 
              colors=['silver','pink','beige','mediumaquamarine','mediumpurple','lavender'])

plt.xlabel('Month Number')
plt.ylabel('Sales unints in Number')
plt.title('Alll product sales data using stack plot')
plt.legend(loc='upper left')
plt.show()
```
